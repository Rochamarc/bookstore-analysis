```python
import pandas as pd
```

```python
import json
```

```python
file = None

with open('books.json', 'r') as f:
    file = json.load(f)
```

```python
df = pd.DataFrame(file).set_index('title')
```

```python
# Remove the urls

df = df.drop(['url','author','format','publisher'], axis=1)
```

```python
# The main dataframe

df
```

```python
authors = df['author_name'].value_counts()[:20]
```

```python
# List of autores with more than one book 

authors
```

```python
# A graphical representation

authors.plot(kind='barh')
```

```python
# Series of author's nationality

nationality = df['author_nationality'].value_counts()[:20]
```

```python
nationality
```

```python
nationality.plot(kind='bar')
```

```python
publishers = df['publisher_name'].value_counts()[:20]
```

```python
# Series with the publishers

publishers
```

```python
# Graphical representation of publishers

publishers.plot(kind='barh')
```

```python
publication = df['published_at'].value_counts()[:20]
```

```python
publication.plot(kind='barh')
```
